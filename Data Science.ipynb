```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, WeekdayLocator
from datetime import timedelta
%matplotlib inline
```

```python
df = pd.read_csv(
    'https://covid.ourworldindata.org/data/owid-covid-data.csv', 
    usecols=['date', 'location', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred'], 
    parse_dates=['date'])
    
countries = ['United States', 'Canada', 'United Kingdom', 'Germany', 'Israel']
df = df[df['location'].isin(countries)]
```

```python
main_country = 'United States'
colors = {country:('grey' if country!= main_country else '#129583') for country in countries}
alphas = {country:(0.75 if country!= main_country else 1.0) for country in countries}
```

```python
# Plot total vaccinations per hundred
pivot = pd.pivot_table(
    data=df,
    index='date',
    columns='location',
    values='total_vaccinations_per_hundred',
    aggfunc='mean',
    )

pivot = pivot.fillna(method='ffill')

fig, ax = plt.subplots(figsize=(12,8))
fig.patch.set_facecolor('#F5F5F5')
ax.patch.set_facecolor('#F5F5F5')

for country in countries:
    ax.plot(
        pivot.index,
        pivot[country],
        color=colors[country],
        alpha=alphas[country]
    )
    ax.text(
        x = pivot.index[-1] + timedelta(days=2),
        y = pivot[country].max(),
        color = colors[country],
        s = country,
        alpha=alphas[country]
    )
plt.ylabel('Total Vaccinations per 100 People', fontsize=12, alpha=0.9)
plt.xlabel('Date', fontsize=12, alpha=0.9)
plt.title('COVID-19 Vaccinations Internationally', fontsize=18, weight='bold', alpha=0.9)
```

```python
# Plot people vaccinated per hundred
pivot = pd.pivot_table(
    data=df,
    index='date',
    columns='location',
    values='people_vaccinated_per_hundred',
    aggfunc='mean',
    )

pivot = pivot.fillna(method='ffill')

fig, ax = plt.subplots(figsize=(12,8))
fig.patch.set_facecolor('#F5F5F5')
ax.patch.set_facecolor('#F5F5F5')

for country in countries:
    ax.plot(
        pivot.index,
        pivot[country],
        color=colors[country],
        alpha=alphas[country]
    )
    ax.text(
        x = pivot.index[-1] + timedelta(days=2),
        y = pivot[country].max(),
        color = colors[country],
        s = country,
        alpha=alphas[country]
    )
plt.ylabel('People Vaccinated per 100 People', fontsize=12, alpha=0.9)
plt.xlabel('Date', fontsize=12, alpha=0.9)
plt.title('COVID-19 Vaccinations Internationally', fontsize=18, weight='bold', alpha=0.9)
```
